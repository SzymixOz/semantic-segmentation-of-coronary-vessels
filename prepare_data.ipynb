```python
import os
from random import random 

image_names = os.listdir("binary_mask")
train_test_split = open("train_test_split.txt", 'w')
for i, image_name in enumerate(image_names):
    if random() < 0.8:
        train_test_split.write(image_name + ' 1\n')
    else:
        train_test_split.write(image_name + ' 0\n')
```

```python
import os

os.mkdir("images_train")
os.mkdir("images_train/input")
os.mkdir("images_train/output")

os.mkdir("images_test")
os.mkdir("images_test/input")
os.mkdir("images_test/output")
```

```python
import shutil

train_test_split = open("train_test_split.txt", 'r')

for line in train_test_split:
    image_name, train_or_test = line.split()
    if train_or_test == '1':
        shutil.copy("binary_mask/" + image_name, "images_train/input/" + image_name)
        shutil.copy("segmentation_mask/" + image_name, "images_train/output/" + image_name)
    else:
        shutil.copy("binary_mask/" + image_name, "images_test/input/" + image_name)
        shutil.copy("segmentation_mask/" + image_name, "images_test/output/" + image_name)
```
